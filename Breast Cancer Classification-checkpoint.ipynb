```python
# dropping unnecessary columns
df.drop(columns =["id","Unnamed: 32"], inplace = True)
```

```python
# removing perimeter_mean,area_mean,perimeter_worst and perimeter_se
df.drop(columns=["perimeter_mean","area_mean","perimeter_worst","perimeter_se","area_worst","area_se"],inplace = True)
```

```python
# checking for outliers

features = df.drop(columns = ["diagnosis"])
fig,[ax1,ax2,ax3,ax4,ax5,ax6] = plt.subplots(6,4)
fig.set_figheight(75)
fig.set_figwidth(30)

for i in range(0,4):
    ax1[i].boxplot(features[features.columns[i]])
    ax1[i].set_xlabel(features.columns[i],fontsize = 20)
    
    ax2[i].boxplot(features[features.columns[i+4]])
    ax2[i].set_xlabel(features.columns[i+4],fontsize = 20)
    
    ax3[i].boxplot(features[features.columns[i+8]])
    ax3[i].set_xlabel(features.columns[i+8],fontsize = 20)
    
    ax4[i].boxplot(features[features.columns[i+12]])
    ax4[i].set_xlabel(features.columns[i+12],fontsize = 20)
    
    ax5[i].boxplot(features[features.columns[i+16]])
    ax5[i].set_xlabel(features.columns[i+16],fontsize = 20)
    
    ax6[i].boxplot(features[features.columns[i+20]])
    ax6[i].set_xlabel(features.columns[i+20],fontsize = 20)

    
```

```python
# removing outliers
df.drop(df[df["radius_se"] > 2].index, inplace = True)
df.drop(df[df["concavity_se"] > 0.2].index, inplace = True)
df.drop(df[df["fractal_dimension_worst"] > 2].index, inplace = True)
```

```python
# converting categorical values into binary values
from sklearn.preprocessing import LabelBinarizer
binarizer = LabelBinarizer()
Y = binarizer.fit_transform(df["diagnosis"]).reshape(565,)
x = df.drop(columns =["diagnosis"]).values
```

```python
# splitting data into train and test sets
from sklearn.model_selection import train_test_split
split = train_test_split(x,Y,test_size = 0.2,shuffle = True, stratify = Y,random_state = 100)
x_train,x_test,y_train,y_test = split
```

```python
# using random forest classifier
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 75, max_depth = 50)
classifier.fit(x_train,y_train)

report = classification_report(y_test, classifier.predict(x_test), target_names = ["B","M"])
print(report)
```
